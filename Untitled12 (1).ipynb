```python


import matplotlib.pyplot as plt
import numpy as np

# Sample data (replace with your actual runtime data)
algorithms = ['Algorithm A', 'Algorithm B', 'Algorithm C']
runtimes = [0.5, 1.2, 0.8]  # Example runtimes in seconds

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(algorithms, runtimes, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel("Algorithms")
plt.ylabel("Runtime (seconds)")
plt.title("Runtime Comparison of Algorithms")

# Add labels above the bars
for i, v in enumerate(runtimes):
    plt.text(i, v + 0.05, f"{v:.2f}", ha='center', va='bottom') # Adjust 0.05 as needed


plt.show()


# Sample data for a line chart (e.g., runtime vs. input size)
input_sizes = np.arange(10, 101, 10)  # Input sizes
runtimes_algA = 0.001 * input_sizes**2  # Example runtime for algorithm A
runtimes_algB = 0.01 * input_sizes * np.log(input_sizes)  # Example runtime for algorithm B

# Create the line chart
plt.figure(figsize=(8, 6))
plt.plot(input_sizes, runtimes_algA, label='Algorithm A', marker='o', linestyle='-')
plt.plot(input_sizes, runtimes_algB, label='Algorithm B', marker='s', linestyle='--')

plt.xlabel("Input Size")
plt.ylabel("Runtime (seconds)")
plt.title("Runtime vs. Input Size")
plt.legend()
plt.grid(True)
plt.show()
```
